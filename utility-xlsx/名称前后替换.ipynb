```python
import os

# 输入文件夹路径
folder_path = r'C:\Users\GL2\Desktop\2025届毕业生就业意向摸底数据'  # 替换为你的文件夹路径

# 遍历文件夹中的所有文件
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):  # 只处理 Excel 文件
        # 通过下划线拆分文件名
        parts = filename.split('_')  # 按下划线分割文件名
        
        # 判断是否符合我们期望的结构（至少两个部分）
        if len(parts) > 1:
            # 提取文件名前部分和学院部分（后部分）
            # 前部分通常是数据名称，后部分是学院名（假设最后一部分是学院名）
            first_part = '_'.join(parts[:-1])  # 前部分
            last_part = parts[-1].replace(".xlsx", "")  # 后部分，去除扩展名

            # 交换前后部分，生成新的文件名
            new_filename = f"{last_part}_{first_part}.xlsx"

            # 生成完整的旧文件路径和新文件路径
            old_file = os.path.join(folder_path, filename)
            new_file = os.path.join(folder_path, new_filename)

            # 重命名文件
            os.rename(old_file, new_file)
            print(f"Renamed: {filename} -> {new_filename}")

print("所有文件已成功重命名！")
```
